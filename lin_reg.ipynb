```python


import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
arr_x = np.array([1.0, 1.6, 3.4, 4.0, 5.2])
arr_y = np.array([1.2, 2.0, 2.4, 3.5, 3.5])

n = arr_x.size

sum_x = np.sum(arr_x)
sum_y = np.sum(arr_y)
sum_xy = np.sum(arr_x*arr_y)
sum_x2 = np.sum(arr_x**2)

den = n * sum_x2-(sum_x)**2
a = (n*sum_xy-(sum_x*sum_y))/den
b = (sum_y*sum_x2-(sum_xy*sum_x)) / den

y = a + b*arr_x
print(y)
```

```python
fig, ax = plt.subplots(figsize=(5, 4), layout='constrained')
ax.scatter(arr_x, arr_y, s=50, facecolor='r', edgecolor='r')  # Plot some data on the axes.
plt.plot(arr_x, y, label='y=.54+.87x')  # Plot some data on the (implicit) axes.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()
```
